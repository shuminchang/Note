# 直線圖

```python
fmri = sns.load_dataset("fmri")
ax = sns.lineplot(x="timepoint", \
                  y="signal", \
                  hue="event", \
                  style="event", \
                  err_style="bars", \
                  ci=100, \
                  markers=False, \
                  dashes=False, \
                  units="subject", \
                  estimator=None, \
                  data=fmri
                 )
plt.xlabel("jwojef", fontsize=13)  # 可改寫X軸label, fontsize: 字型大小
```

- x, y: data for x, y-axis，會自動顯示label
- hue: 顏色分類
- style: 圖的類型（ex.線的類型）
- err_style: error bar的類型
- ci: error bar的長度（0-100，數字愈大愈長）
- markers: 折線顯示標記
- dashes: 虛線開關
- unit: 分類後畫出每個分類的直線圖
- estimator: 與unit搭配
- data: data

# distplot

```python
from scipy.stats import norm
sns.set(); np.random.seed(0)
x = np.random.randn(100)
ax = sns.distplot(x, vertical=False, fit=norm, kde=False, color="y")

# 計算平均數跟標準差
(mu, sigma) = norm.fit(x)
print('mu = {:.2f} and sigma = {:.2f}\n'.format(mu, sigma))
plt.legend(['Normal dist. ($\mu=$ {:.2f} and $\sigma=$ {:.2f})'.format(mu, sigma)], loc='best')
```

- vertical: True - 把直方圖變橫的
- fit=normal: 加入normal distribution
- kde: kde開關
- color: 直方圖顏色

# barplot

```python
from numpy import median
sns.set(style='whitegrid')
tips = sns.load_dataset('tips')
ax = sns.barplot(x='day', \
                 y='total_bill', \
                 hue='sex', \
                 order=["Sun", "Sat", "Thur", "Fri"], \
                 ci="sd", \
                 capsize=.2, \
                 palette="Blues_d", \
                 dodge=False, \
                 saturation=.5, \
                 data=tips, \
                 estimator=median
                )
```

- 會自動顯示error bar
- x, y軸的資料對調會改變barplot的方向
- order: 依照order的list安排barplot的資料順序
- estimator: 估計方式
- ci="sd": 顯示標準差，default為信心區間
- capsize: error bar加上"caps"
- palette: 改變顏色
- dodge=False: hue區分的資料不並列顯示，而是顯示在同一個位置
- saturation: 飽和度？

# Heatmaps
## 依照數值大小給予顏色區分

```python
# Load the example flights dataset and conver to long-form
flights_long = sns.load_dataset("flights")
# 將data依"month", "year"分類
flights = flights_long.pivot("month", "year", "passengers")

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(flights, annot=True, fmt="d", linewidths=.5, ax=ax)
```

- annot: 顯示數值
- fmt="d": 不設定會以科學記號的型式顯示
- linewidth: 方格間的白線寬度
